```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"TITANIC.csv")
df
```

```python
df.info()
```

```python
(df.isnull().sum()/df.shape[0])*100
```

```python
df["Age"].value_counts()
```

```python
df["Age"].isnull().sum()
```

```python
df["Age"].mode()[0]
```

```python
df["Age"].fillna(df["Age"].mode()[0],inplace=True)
df
```

```python
df.dtypes
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
df['Cabin'].isnull().sum()
df
```

```python
df['Fare'].fillna(df['Fare'].mode()[0],inplace=True)
df['Fare'].isnull().sum()
df
```

```python
sns.heatmap(df.isnull())
plt.show()
```

```python
plt.hist(df['Survived'], bins=2, edgecolor='black')
plt.xlabel('Survived')
plt.ylabel('Frequency')
plt.title('Histogram of Survived Column')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# Survival rate by gender
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean()
print(survival_rate_by_gender)
```

```python
survival_rate_by_gender.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Survival rate by passenger class
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()
print(survival_rate_by_class)
```

```python
survival_rate_by_class.plot(kind='bar')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='Age', by='Survived')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.title('Box Plot of Age by Survival Status')
plt.suptitle('')
plt.show()
```
